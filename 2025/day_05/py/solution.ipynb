```python
import sys
mode = "real"
#mode = "test"
filename = None
```

```python
if mode == "real":
    with open(filename, 'r') as file:
        raw = file.read()
else:
    raw = "3-5\n10-14\n16-20\n12-18\n\n1\n5\n8\n11\n17\n32\n"
```

```python
class interval:
    def __init__(self, definition: str):
        self.start, self.end = map(int, definition.split("-"))

    def __len__(self):
        return self.end - self.start + 1
    
    def __lt__(self, other):
        return self.start < other.start
    
    def __contains__(self, other) -> bool:
        """Check if two intervals overlap, or if an int is """
        if isinstance(other, interval):
            return self.start <= other.end and other.start <= self.end
        if isinstance(other,int):
            return self.start <= other <= self.end
        return False
        
    def __add__(self, other):
        if not other in self:
            raise ValueError("Cannot add non-overlapping intervals")
        return interval(f"{min(self.start, other.start)}-{max(self.end, other.end)}")
```

```python
ranges, values = map(str.splitlines, raw.split("\n\n"))
intervals = sorted([interval(r) for r in ranges])
values = list(map(int, values))
```

```python
new_ints = []
last_merged = None
for itv in intervals:
    if last_merged in itv:
        last_merged +=itv
    else:
        if last_merged:
            new_ints.append(last_merged)
        last_merged = itv
new_ints.append(last_merged)
```

```python
print(f"Solution 1: {sum(any(value in i for i in new_ints) for value in values)}")
print(f"Solution 2: {sum(len(i) for i in new_ints)}")
```
