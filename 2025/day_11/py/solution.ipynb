```python
import networkx as nx
G = nx.DiGraph()
for row in raw:
    start,ends = row.split(":")
    if not G.has_node(start):
        G.add_node(start)
    for node in ends.split(" "):
        if not G.has_node(node):
            G.add_node(node) 
        G.add_edge(start,node)
```

```python
def cpath(a, b):
    topo = list(nx.topological_sort(G))
    counts = {n: 0 for n in topo}
    counts[a] = 1
    for u in topo:
        for v in G.successors(u):
            counts[v] += counts[u]
    return counts[b]
```
