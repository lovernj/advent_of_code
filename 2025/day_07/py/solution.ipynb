```python
from collections import defaultdict
mode = "real"
#mode = "test"
filename = r"filename"
```

```python
if mode == "real":
    with open(filename, "r") as f:
        lines = f.read().splitlines()
else:
    lines = ".......S.......\n...............\n.......^.......\n...............\n......^.^......\n...............\n.....^.^.^.....\n...............\n....^.^...^....\n...............\n...^.^...^.^...\n...............\n..^...^.....^..\n...............\n.^.^.^.^.^...^.\n...............".split()
```

```python
paths = {lines[0].find("S")}
splits = 0
for row in lines[::2]:
    nextpaths = set()
    for path in paths:
        if row[path] == "^":
            splits += 1
            if path-1 not in paths:
                nextpaths.add(path-1)
            if path+1 not in paths:
                nextpaths.add(path+1)
        else:    
            nextpaths.add(path)
    paths = nextpaths
print("Solution 1:",splits)
```

```python
paths = {lines[0].find("S"): 1}
for row in lines[::2]:
    nextpaths = defaultdict(int)
    for idx in paths.keys():
        if row[idx] == "^": # split
            nextpaths[idx-1] += paths[idx]
            nextpaths[idx+1] += paths[idx]
        else:
            nextpaths[idx] += paths[idx]
    paths = nextpaths
print("Solution 2:",sum(paths.values()))
```

```python
import functools
@functools.lru_cache()
def getTimelines(x:int,y:int)-> int:
    if y==len(lines)-1:
        return 1
    if lines[y][x]=="^":
        return getTimelines(x+1,y)+getTimelines(x-1,y)
    return getTimelines(x,y+1)
print("Solution 2:",getTimelines(lines[0].find("S"),0))
```
