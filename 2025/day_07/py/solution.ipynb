```python
from collections import defaultdict
mode = "real"
#mode = "test"
filename = r"filename"
```

```python
if mode == "real":
    with open(filename, "r") as f:
        lines = f.read().splitlines()
else:
    lines = ".......S.......\n...............\n.......^.......\n...............\n......^.^......\n...............\n.....^.^.^.....\n...............\n....^.^...^....\n...............\n...^.^...^.^...\n...............\n..^...^.....^..\n...............\n.^.^.^.^.^...^.\n...............".split()
```

```python
lines[2::2]
```

```python
paths = {lines[0].find("S")}
splits = 0
for row in lines[::2]:
    nextpaths = set()
    for path in paths:
        if row[path] == "^":
            splits += 1
            if path-1 not in paths:
                nextpaths.add(path-1)
            if path+1 not in paths:
                nextpaths.add(path+1)
        else:    
            nextpaths.add(path)
    paths = nextpaths
print("Solution 1:",splits)
```

```python
paths = {lines[0].find("S"): 1}
for row in lines[::2]:
    nextpaths = defaultdict(int)
    for idx in paths.keys():
        if row[idx] == "^": # split
            nextpaths[idx-1] += paths[idx]
            nextpaths[idx+1] += paths[idx]
        else:
            nextpaths[idx] += paths[idx]
    paths = nextpaths
print(sum(paths.values()))
```

```python
s = lines[0].index("S")

# part 1 (count splits)
splits = 0
paths = {s}
for row in lines[1:]:
    splits += sum(1 for p in paths if row[p] == "^")
    paths = ({p-1 for p in paths if row[p] == "^" and p-1 not in paths}
             | {p+1 for p in paths if row[p] == "^" and p+1 not in paths}
             | {p for p in paths if row[p] != "^"})
print("Solution 1:", splits)
```
